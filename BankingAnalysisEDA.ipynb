```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
query = "SELECT * FROM banking.customer"

df = pd.read_sql(query, con)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# Generating descriptive statistics for the dataframe.
df.describe()
```

```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Medium', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins = bins, labels = labels, right = False)
```

```python
df['Income Band'].value_counts().plot(kind = "bar")
```

```python
# Examine the distribution of unique categories in categorical columns.
categorical_cols = df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]]

for col in categorical_cols:
    print(f"Value Counts for '{col}':")
    display(df[col].value_counts()).pol
```

```python
for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]]):
    plt.figure(i)
    sns.countplot(data = df, x = predictor, hue = 'GenderId')
```

```python
for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]]):
    plt.figure(i)
    sns.countplot(data = df, x = predictor, hue = 'Nationality')
```

```python
# Histogram plot of value counts for different Occupation.

for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title("Histogram of Occupation Count")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

# Numerical Analysis

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance','Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts']

#univariate analysis and visualization
plt.figure(figsize=(15,10))
for i,col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde = True)
    plt.title(col)
plt.show()
```

# Heatmaps

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance','Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts']

correlation_matrix = df[numerical_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt = ".2f")
plt.title("Correlation Matrix")
plt.show()
```
